```python
nifty_future=nifty_future.loc[(nifty_future.index.time>=datetime.time(9,15)) & (nifty_future.index.time<=datetime.time(15, 30)),:]
nifty_future.dropna(inplace=True)
```

```python
nifty=nifty.loc[(nifty.index.time>=datetime.time(9,15)) & (nifty.index.time<=datetime.time(15, 30)),:]
nifty.dropna(inplace=True)
```

```python
buy_nifty=nifty.Close[0]
sell_nifty_Fut=nifty_future.Close[0]
Difference=sell_nifty_Fut-buy_nifty
```

```python
Brokerage_Fut=195.19

Brokerage_equity=1953.36 
```

```python
Total_brokerage=Brokerage_Fut+Brokerage_equity
Capital_req=buy_nifty*75
interest_rate=.05
```

```python
for i in range(1,nifty.shape[0]):
    interest_charge=interest_rate*Capital_req/365*(nifty.index[i].day)
    total_cost=interest_charge+Total_brokerage
    if (Difference-(nifty_future.loc[nifty.index[i],"Close"]-nifty.Close[i]))*75>total_cost:
        print("Arbitrage existed at",nifty.index[i], (nifty_future.Close[i]-nifty.Close[i]))
    else:
        print("arb does not existed at",nifty.index[i], (nifty_future.Close[i]-nifty.Close[i]))
```
